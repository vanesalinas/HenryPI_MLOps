```python
import pandas as pd
import numpy as np
import gzip
import json
import ast
from pandas import json_normalize
```

EXTRACT (EXTRACCION)

Extraemos los datos desde los archvos JSON proporcionados

```python
# Lista para almacenar los datos del archivo JSON
fila = []

# Abrir el archivo comprimido con gzip y leer el contenido
with gzip.open('../Datasets/steam_games.json.gz', 'rt') as games:
    for linea in games.readlines():
        # Cargar cada línea como un objeto JSON
        fila.append(json.loads(linea)) 

# Crear un DataFrame desde la lista de diccionarios
steam_games = pd.DataFrame(fila) 

steam_games
```

Nos encontramos con datos anidados dentro de los dataframes por lo que creamos una funcion para desanidar dichos datos.

```python
from funcion_desanidar import descomprimirjson
```

```python
# Lista para almacenar los datos del archivo JSON
fila_reviews = []

# Abrir el archivo comprimido con gzip y leer el contenido
with gzip.open('../Datasets/user_reviews.json.gz', 'rt', encoding='MacRoman') as reviews:
    for linea in reviews:
        # Cargar cada línea como un objeto JSON
        fila_reviews.append(ast.literal_eval(linea)) 

# Crear un DataFrame desde la lista de diccionarios
user_reviews = pd.DataFrame(fila_reviews) 

user_reviews
```

```python
# Lista para almacenar los datos del archivo JSON
fila_items = []

# Abrir el archivo comprimido con gzip y leer el contenido
with gzip.open('../Datasets/users_items.json.gz', 'rt', encoding='MacRoman') as items:
    for linea in items:
        # Cargar cada línea como un objeto JSON
        fila_items.append(ast.literal_eval(linea)) 

# Crear un DataFrame desde la lista de diccionarios
user_reviews = pd.DataFrame(fila_items) 

user_reviews
```
